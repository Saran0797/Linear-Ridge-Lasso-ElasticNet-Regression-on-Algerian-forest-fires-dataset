# Linear Regression Model Fitting

```python
from sklearn.linear_model import LinearRegression
```

```python
linear = LinearRegression()
```

```python
#fitting x_train and y_train 
linear.fit(x_train, y_train)
```

```python
#co-efficient of x_train and y_train
linear.coef_
```

```python
#intercept of x_train and y_train
linear.intercept_
```

```python
#predicting the target variable by using x_test
y_pred = linear.predict(x_test)
y_pred
```

```python
#finding the residuals of the dataset 
residual = y_test-y_pred
residual
```

```python
#to check the assumptions of linear model
#y_test and y_pred should be normally distributed

sns.regplot(y_test, y_pred)
```

```python
from scipy.stats import probplot
```

```python
#2nd assumption
#residuals are normally distributed

plt.figure(figsize=(7,5))

plt.subplot(1,2,1)
plt.title('distribution of residuals')
sns.distplot(residual)

plt.subplot(1,2,2)
plt.title('QQ-plot of residuals')
probplot(residual, plot=plt, fit=True, dist='norm')
plt.show()
```

```python
#3rd assumption
#errors to be spread randomly

sns.scatterplot(x=residual, y=y_pred)
```

# performance metrics 

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import median_absolute_error
```

```python
print (mean_squared_error(y_test, y_pred))
print (mean_absolute_error(y_test, y_pred))
print (np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
#r2 score of the model 
from sklearn.metrics import r2_score
```

```python
#r2 score 
r2_score = r2_score(y_test, y_pred)
r2_score
```

```python
#adjusted r2 score
#Adjusted R2 = 1 â€“ [(1-R2)*(n-1)/(n-k-1)] formula
adj_R2 = 1 - (1-r2_score)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
adj_R2 
```
